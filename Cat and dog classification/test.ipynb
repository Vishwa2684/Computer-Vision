```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# CNN = tf.keras.models.load_model('./CNN.h5')

doggo = cv2.imread('C:\\Users\\vishw\\OneDrive\\Desktop\\maple 1.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(doggo)
```

```python
doggo = cv2.resize(doggo,(50,50))
print(doggo.shape)

plt.imshow(doggo,cmap='gray')
```

```python
doggo =doggo.reshape((50,50,1))
doggo = np.expand_dims(doggo,axis=0)
```

```python
CNN = tf.keras.models.load_model('./CatsAndDogs.h5')
```

```python
y_pred = CNN.predict(doggo)
```

```python

print(y_pred)
    
if y_pred[0][0] > 0.5:
        print("Prediction: Dog")
else:
        print("Prediction: Cat")
```
