```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python

doggo = cv2.imread('C:\\Users\\vishw\\OneDrive\\Desktop\\maple x.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(doggo,cmap='gray')
```

```python
doggo = cv2.resize(doggo,(50,50))
print(doggo.shape)

plt.imshow(doggo,cmap='gray')
```

```python
doggo =doggo.reshape((50,50,1))
doggo = np.expand_dims(doggo,axis=0)
```

```python
cat_img = cv2.imread('C:\\Users\\vishw\\OneDrive\\Desktop\\images.jpg',cv2.IMREAD_GRAYSCALE)
print(cat_img.shape)
plt.imshow(cat_img,cmap='gray')
```

```python
cat_img = cv2.resize(cat_img,(50,50))
cat_img = cat_img.reshape((50,50,1))
plt.imshow(cat_img,cmap='gray')

cat_img = np.expand_dims(cat_img,axis=0)
```

```python
# Function to get result
def get_result(y_pred):
    if y_pred[0][0] > 0.5:
            return 'Dog'
    else:
            return 'Cat'
```

```python
CNN = tf.keras.models.load_model('./CatsAndDogs.h5')
```

```python
dog_pred = CNN.predict(doggo)
get_result(dog_pred)
```

```python
cat_pred =CNN.predict(cat_img)
get_result(cat_pred)
```
