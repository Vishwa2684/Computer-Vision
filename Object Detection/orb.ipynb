```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
face1 = cv2.imread('./inputs/face.jpg')
face2 = cv2.imread('./inputs/face4.jpg')
```

```python
orb = cv2.ORB_create()
kp1,desc1 = orb.detectAndCompute(face1,None)

kp2,desc2 = orb.detectAndCompute(face2,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING2,crossCheck=True)

matches = bf.match(desc1,desc2)
 
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

matched_img = cv2.drawMatches(face1,kp1,face2,kp2,matches[:25],None,flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)

plt.imshow(matched_img)

cv2.imwrite('./results/orb_matching.jpg',matched_img)
```
